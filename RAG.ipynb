```python
from transformers import pipeline

# Load a language model pipeline (let's use a small GPT-like model)
model = pipeline("text-generation", model="gpt2")

# Ask the model a question
question = "What is the capital of France?"

# The model tries to answer the question from its own memory
response = model(question, max_length=30, num_return_sequences=1)

# Print the answer
print("Answer without RAG:", response[0]['generated_text'])
```

```python
# with rag
from transformers import RagRetriever, RagTokenizer, RagSequenceForGeneration
!pip install datasets
!pip install faiss
```

```python
tokenizer = RagTokenizer.from_pretrained("facebook/rag-sequence-nq")
retriever = RagRetriever.from_pretrained("facebook/rag-sequence-nq")
model = RagSequenceForGeneration.from_pretrained("facebook/rag-sequence-nq", retriever=retriever)


# Tokenize the input question
question = "What is the capital of France?"
inputs = tokenizer([question], return_tensors="pt")

# The model first retrieves information and then generates an answer
generated = model.generate(input_ids=inputs["input_ids"], num_beams=2, num_return_sequences=1)

# Decode the generated answer
answer = tokenizer.decode(generated[0], skip_special_tokens=True)
print("Answer with RAG:", answer)
```
